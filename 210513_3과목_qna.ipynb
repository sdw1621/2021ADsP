### 2.편향(bias)-분산(variance) tradeoff
- **지도 학습 알고리즘**이 트레이닝 셋의 범위를 넘어 지나치게 일반화 하는 것을 예방하기 위해 두 종류의 오차(편향, 분산)을 최소화 할 때 겪는 문제
- 편향(bais) : 학습 알고리즘에서 잘못된 가정을 했을 때 발생하는 오차, 높은 편향값은 과소적합(underfitting) 문제를 발생시킨다.
- 분산(variance) : 트레이닝 셋에 내제된 작은 변동 때문에 발생하는 오차로 높은 분산값은 큰 노이즈까지 모델링에 포함시키는 과적합(overfitting) 문제를 발생시킨다.
- 요약
   - 과소적합 문제가 발생한 경우 편향(bias)이 크다고 표현함
   - 과대적합 문제가 발생한 경우 분산(variance)가 크다고 표현함
   - bias와 variance는 tradeoff 관계임
